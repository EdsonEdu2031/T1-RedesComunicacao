```python
# Bibliotecas para importar horário
from datetime import datetime
import pytz
import numpy as np
import random
```

```python
# Variáveis globais

acoes = {"PETR4": 39.67, "VALE3": 88.73}

carteira = {"PETR4": 2, "VALE3":4}

# Funções
def mostrar_acoes():
  print('\nAções disponíveis: ')
  for acao in acoes:
    print(f'{acao} - R${acoes[acao]:.2f}')
```

# Lógica do processamento dos comandos hardcoded

```python
# Definindo o fuso horário de Brasília
brasiliatime = pytz.timezone('America/Sao_Paulo')

# Pegando a hora atual no fuso certo
hora = datetime.now(brasiliatime).time().replace(microsecond=0)
print(f'{hora} <CONECTADO!!>\n')
print("Bem-vindo ao sistema de Bolsa de Valores\n")

mostrar_acoes()

# Loop para ficar rodando infinitamente
while(True):
  # Pegando a ação
  op = input('\nInforme o comando: ')

  comando = op.split(' ')

  # Remove strings vazias resultantes de múltiplos espaços
  comando = [c for c in comando if c]

  print(f"Você executou: {op}\n")

  # Checando se é um comando válido
  if not comando:
      print('Comando vazio. Por favor, insira um comando.')
      continue

  if comando[0] == 'actions':
    mostrar_acoes()
    continue

  # Se a ação for buy, checamos o formato, se a quantidade é válida (>0) e se a ação digitada existe
  elif comando[0] == 'buy':
    if len(comando) < 3:
        print('Formato do comando BUY inválido. Use: buy <ação> <quantidade>')
    else:
        try:
            acao_nome = comando[1].upper()
            quantidade = float(comando[2])
            if acao_nome in acoes:
                if quantidade > 0:
                    # Ação de compra
                    carteira[acao_nome] = carteira.get(acao_nome, 0) + quantidade
                    print(f'Comprado {quantidade} de {acao_nome}. Sua carteira: {carteira}')
                else:
                    print('Quantidade inválida para compra (deve ser maior que 0)')
            else:
                print(f'Ação {acao_nome} não encontrada na lista de ações disponíveis.')

        except ValueError:
            print('Quantidade inválida. A quantidade deve ser um número.')
    continue

  # Se a ação for sell, checamos o formato, se a ação está na carteira e se quantidade é válida (>0 e <= o que tem na carteira)
  elif comando[0] == 'sell':
    if len(comando) < 3:
        print('Formato do comando SELL inválido. Use: sell <ação> <quantidade>')
    else:
        try:
            acao_nome = comando[1].upper()
            quantidade = float(comando[2])
            if acao_nome in carteira:
                if quantidade > 0 and quantidade <= carteira[acao_nome]:
                    # Ação de venda (improvisada)
                    carteira[acao_nome] -= quantidade
                    if carteira[acao_nome] == 0:
                        del carteira[acao_nome] # Remove se a quantidade for zero
                    print(f'Vendido {quantidade} de {acao_nome}. Sua carteira: {carteira}')
                else:
                    print('Quantidade inválida para venda ou insuficiente na carteira.')
            else:
                print(f'Ação {acao_nome} não encontrada na sua carteira.')
        except ValueError:
            print('Quantidade inválida. A quantidade deve ser um número.')
    continue

  # Se a ação for carteira, checamos o formato, e mostramos as informações dela
  elif comando[0] == 'carteira':
    if not carteira:
      print('Sua carteira está vazia.')
    else:
      print('Ações na carteira: ')
      for acao, qtd in carteira.items():
        print(f'{acao}: {qtd} unidades')
    continue

  # Se a ação for exit, só quebra o loop
  elif comando[0] == 'exit':
    print('Saindo do sistema')
    break

  else:
    print('Comando inválido. Comandos disponíveis: actions, buy, sell, carteira, exit')
    continue
```

```python
# Variador de preços
# Variando eles de -1 a 1 de 0.1 em 0.1
var_precos = np.arange(-1, 1.1, 0.1)
var_precos = var_precos.round(2)
mostrar_acoes()
# Para cada ações irá sortear um valor e incrementar
for i in range(4): # <-- incrementar loop infinito com delay
  for acao in acoes:
    acoes[acao] += random.choice(var_precos)
  mostrar_acoes()
```

```python
def atualizaAcoes(acoes: tuple):
    # Variador de preços
    # Variando eles de -1 a 1 de 0.1 em 0.1
    var_precos = np.arange(-1, 1.1, 0.1)
    var_precos = var_precos.round(2)
    
    for acao in acoes:
        acoes[acao] += random.choice(var_precos)

    return acoes
```
